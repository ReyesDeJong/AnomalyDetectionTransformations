```python
import os
import sys

PROJECT_PATH = os.path.abspath(
    os.path.join(os.path.dirname('__file__'), '..'))
sys.path.append(PROJECT_PATH)
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib.inline
```

```python

def plot_input_image(image, name, aucroc, auroc_std):
    #fill titles with blanks
  fontsize=17
  fig = plt.figure(figsize=(4.5,3))
  plt.imshow(image, interpolation='nearest', vmin=0, vmax=1)
  plt.xticks(fontsize=fontsize)
  plt.yticks(fontsize=fontsize)
  plt.title('%i Transformations\nAUROC: %.2f+/-%.2f'% (len(image),aucroc,auroc_std), fontsize=fontsize)
  plt.xlabel('Transformation index', fontsize=fontsize)
  plt.ylabel('Transformation index', fontsize=fontsize)
  cb =plt.colorbar()
  cb.set_label('Accuracy', fontsize=fontsize)
  for t in cb.ax.get_yticklabels():
   t.set_fontsize(fontsize)
  plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
              hspace = 0, wspace = 0.1)

  plt.margins(0,0)
  plt.savefig(os.path.join(PROJECT_PATH,'figure_creation/figs/%s.svg' % name),
              format='svg', dpi=600, bbox_inches = 'tight', pad_inches = 0,
              transparent=True)
  plt.show()
```

```python
#18MAt
data_path_model = os.path.join(
    PROJECT_PATH, 
    'results/Ensemble_OVO_Transformer_OD_Simple_Model/transform_selection/hits_4_channels')
data_path_18 = os.path.join(data_path_model, '18_Transform/results')
data_path_9 = os.path.join(data_path_model, 'Trans_Transformer/results')
data_path_72 = os.path.join(data_path_model, 'Transformer/results')
```

```python
matrix_path_18 = os.path.join(data_path_18, 'train_acc_matrix.pkl')
matrix_path_9 = os.path.join(data_path_9, 'train_acc_matrix.pkl')
matrix_path_72 = os.path.join(data_path_72, 'train_acc_matrix.pkl')
```

```python
load_train_mat = pd.read_pickle(matrix_path_18)
plot_input_image(load_train_mat, 'disc_mat_18', 98.93, 0.12)

load_train_mat = pd.read_pickle(matrix_path_72)
plot_input_image(load_train_mat, 'disc_mat_72', 98.78, 0.26)

load_train_mat = pd.read_pickle(matrix_path_9)
plot_input_image(load_train_mat, 'disc_mat_9', 99.16, 0.13)
```
